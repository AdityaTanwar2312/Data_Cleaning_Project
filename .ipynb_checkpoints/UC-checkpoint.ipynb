```python
import pandas as pd
import matplotlib
import numpy as np
```

```python
n_values=['-1', 'Unknown / Non-Applicable']

df = pd.read_csv('data/Uncleaned_DS_jobs.csv', na_values=n_values)
```

```python
df.set_index('index', inplace=True)
```

```python
df
```

```python
job_title = df.groupby(['Job Title'])
```

```python
df['Salary Estimate'] = df['Salary Estimate'].str.replace('K', ' ')
```

```python
df[['Low','High']] = df['Salary Estimate'].str.split('-', expand=True)
```

```python
df['High'] = df['High'].str.split('(').str[0]
```

```python
df.drop('Job Description', axis=1, inplace=True)
```

```python
df['Company Name'] = df['Company Name'].str.split('\n').str[0]
```

```python
df['Revenue'] = df['Revenue'].replace('Less than $1 million (USD)', 900000)
```

```python
df['Low'] = df['Low'].str.replace('$', ' ')
df['High'] = df['High'].str.replace('$', ' ')
```

```python
df[['Low', 'High']] = df[['Low','High']].astype(float)*1000
```

```python
df['Revenue'].unique()
```

```python
df['Revenue'] = df['Revenue'].str.replace('$', ' ')
df['Revenue'] = df['Revenue'].str.replace('(USD)', ' ')
df['Revenue'] = df['Revenue'].str.replace('million', '000000')
df['Revenue'] = df['Revenue'].str.replace('billion', '000000000')
```

```python
df['Revenue'] = df['Revenue'].str.replace('+', ' ')
```

```python
df['Revenue'] = df['Revenue'].str.split('to').str[1]
```

```python
#To remove white spaces from between the text

df['Revenue'] = df['Revenue'].str.replace(r'\s+', '', regex=True)
```

```python
df['Revenue'] = df['Revenue'].astype(float)
```

```python
df['Revenue'] = df['Revenue'].interpolate()
```

```python
df['Size'] = df['Size'].str.replace(r'\s+', '', regex=True)
df['Size'] = df['Size'].str.replace('employees', ' ')
```

```python
df[['Min_Emp', 'Max_emp']] = df['Size'].str.split('to', expand=True)
```

```python
df = df.rename(columns = {'Low':'Min_Salary', 'High':'Max_Salary'})
```

```python
df['Revenue'] = df['Revenue'].bfill()
```

```python
df
```

```python
df['Min_Emp'] = df['Min_Emp'].replace('Unknown', np.nan)
df['Min_Emp'] = df['Min_Emp'].str.replace('+', ' ')
```

```python
df['Min_Emp'].unique()
```

```python
df['Min_Emp'] = df['Min_Emp'].astype(float)
```

```python
df.info()
```

```python
df['Max_emp'].unique()
```

```python
df['Max_emp'] = df['Max_emp'].astype(float)
```

```python
df.to_csv('cleaned_data.csv', index=False)
```
